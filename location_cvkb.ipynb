```python
import pandas as pd
dfs=pd.read_html("https://en.wikipedia.org/wiki/List of postal codes of Canada: M",header=0)
for df in dfs:
    print(df)
df1=dfs[0]
df2=df1.drop(df1.index[df1['Borough']=='Not assigned'])
df2
```

```python
df2.loc[df1['Neighbourhood'] == 'Not assigned', ['Neighbourhood']] = df2.Borough
df2
```

```python
df3 = df2.groupby(['Postcode','Borough'])['Neighbourhood'].apply(', '.join).reset_index()
df3
```

```python
df4=pd.read_csv("http://cocl.us/Geospatial_data")
df4
df5=df4.rename(columns={"Postal Code":"Postcode"})
df5
```

```python
df6=pd.merge(df3,df5,on='Postcode')
df6
```

```python
neigh_toronto=df6
```

```python
address = 'Toronto'

geolocator = Nominatim(user_agent="foursquare_agent")
location = geolocator.geocode(address,timeout=10)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto City are {}, {}.'.format(latitude, longitude))
```

```python
# create map of New York using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(neigh_toronto['Latitude'], neigh_toronto['Longitude'], neigh_toronto['Borough'], neigh_toronto['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
